```python
import svgwrite

def generate_grid(rows, cols, spacing=40, offset=(20, 20)):
    ox, oy = offset
    return [(ox + j * spacing, oy + i * spacing) for i in range(rows) for j in range(cols)]
```

```python
def add_loop(dwg, p_tl, p_tr, p_br, p_bl, stroke_width=2):
    # Simple loop around 4 corner dots
    path = dwg.path(d=f"M {p_tl[0]},{p_tl[1]}", stroke="black", fill="none", stroke_width=stroke_width)
    path.push(f"L {p_tr[0]},{p_tr[1]}")
    path.push(f"L {p_br[0]},{p_br[1]}")
    path.push(f"L {p_bl[0]},{p_bl[1]}")
    path.push("Z")
    dwg.add(path)
```

```python
import random

def generate_kolam(rows=5, cols=5, spacing=50, filename="kolam.svg"):
    pts = generate_grid(rows, cols, spacing)
    dwg = svgwrite.Drawing(filename, size=(cols * spacing + 60, rows * spacing + 60))

    # Draw dots
    for x, y in pts:
        dwg.add(dwg.circle(center=(x, y), r=3, fill="black"))

    # Draw motifs (loops) around random 2x2 cells
    for i in range(rows - 1):
        for j in range(cols - 1):
            p_tl = pts[i * cols + j]
            p_tr = pts[i * cols + j + 1]
            p_bl = pts[(i + 1) * cols + j]
            p_br = pts[(i + 1) * cols + j + 1]
            if random.random() < 0.8:  # 80% chance to draw
                add_loop(dwg, p_tl, p_tr, p_br, p_bl)

    dwg.save()
    print(f"Kolam saved to {filename}")
```

```python
# kolam_dot_detect.py
# Requires: pip install opencv-python numpy matplotlib

import cv2
import numpy as np
import os
from pathlib import Path
import matplotlib.pyplot as plt

# --------------------------------------------------
# Step 1: Create synthetic sample (for testing only)
# --------------------------------------------------
def create_synthetic_sample(out_path="kolam_sample.jpg",
                            rows=6, cols=6, spacing=80, offset=(60,60),
                            dot_radius=6, img_size=(600,600)):
    """Create a synthetic dot-grid image for testing."""
    w, h = img_size
    bg = 255 * np.ones((h, w), dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            x = offset[0] + j * spacing
            y = offset[1] + i * spacing
            cv2.circle(bg, (x, y), dot_radius, (0,), -1)
    cv2.imwrite(out_path, bg)
    return out_path

# --------------------------------------------------
# Step 2: Detect dots
# --------------------------------------------------
def detect_dots_safe(img_path, min_area=5, max_area=10000, debug=False):
    """Load an image and detect dot centroids using adaptive threshold + contours."""
    if not Path(img_path).exists():
        raise FileNotFoundError(f"Image not found: {img_path}\n(Current dir: {os.getcwd()})")

    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"cv2.imread failed to load image: {img_path}")

    # Blur and adaptive threshold
    blur = cv2.GaussianBlur(img, (5,5), 0)
    thr = cv2.adaptiveThreshold(
        blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
        cv2.THRESH_BINARY_INV, 11, 2
    )

    # Morphological opening to clean small noise
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
    clean = cv2.morphologyEx(thr, cv2.MORPH_OPEN, kernel)

    # Find contours (candidate dots)
    contours, _ = cv2.findContours(clean, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    centers = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area < min_area or area > max_area:
            continue
        M = cv2.moments(cnt)
        if M['m00'] == 0:
            continue
        cx = M['m10'] / M['m00']
        cy = M['m01'] / M['m00']
        centers.append((cx, cy))

    centers = np.array(centers)

    if debug:
        print(f"Contours found: {len(contours)}; Dots after filter: {len(centers)}")
        cv2.imwrite("debug_cleaned.png", clean)  # Debug output

    return img, centers, clean

# --------------------------------------------------
# Step 3: Arrange into grid
# --------------------------------------------------
def arrange_grid(points, max_rows=40):
    """Arrange detected points into a grid by grouping rows."""
    if len(points) == 0:
        return None, 0, 0

    pts = np.array(points)
    N = len(pts)
    best_score = 1e9
    best_groups = None

    for r in range(1, min(N, max_rows) + 1):
        sorted_idx = np.argsort(pts[:,1])  # sort by y
        groups = np.array_split(pts[sorted_idx], r)
        if any(len(g) < 2 for g in groups):
            continue
        stds = [np.std(g[:,1]) for g in groups]
        score = np.mean(stds)
        if score < best_score:
            best_score, best_groups = score, groups

    if best_groups is None:
        best_groups = [pts[np.argsort(pts[:,1])]]

    grid = [g[np.argsort(g[:,0])] for g in best_groups]
    rows = len(grid)
    cols = int(round(np.mean([len(r) for r in grid])))
    return grid, rows, cols

# --------------------------------------------------
# Step 4: Draw loops
# --------------------------------------------------
def draw_loops_on_image(img_gray, centers, out_path="kolam_loops.png"):
    """Draw ellipses around 2x2 dot cells."""
    canvas = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR)
    grid, rows, cols = arrange_grid(centers)

    if grid is None:
        print("⚠️ No grid could be formed (no dots).")
        return None

    max_cols = max(len(r) for r in grid)
    pts_grid = [[None]*max_cols for _ in range(len(grid))]
    for i, r in enumerate(grid):
        for j, p in enumerate(r):
            pts_grid[i][j] = (int(round(p[0])), int(round(p[1])))

    for i in range(len(grid)-1):
        for j in range(max_cols-1):
            p_tl = pts_grid[i][j]
            p_tr = pts_grid[i][j+1]
            p_bl = pts_grid[i+1][j] if i+1 < len(grid) else None
            p_br = pts_grid[i+1][j+1] if i+1 < len(grid) else None
            if None in (p_tl, p_tr, p_bl, p_br):
                continue

            minx = min(p_tl[0], p_bl[0])
            maxx = max(p_tr[0], p_br[0])
            miny = min(p_tl[1], p_tr[1])
            maxy = max(p_bl[1], p_br[1])
            center = ((minx + maxx)//2, (miny + maxy)//2)
            axes = (max(6, (maxx-minx)//2), max(6, (maxy-miny)//2))
            cv2.ellipse(canvas, center, axes, 0, 0, 360, (0,0,255), 2)

    cv2.imwrite(out_path, canvas)
    return out_path

# --------------------------------------------------
# Step 5: Main
# --------------------------------------------------
def main():
    sample_path = "kolam_sample.jpg"

    # Create synthetic test image if not present
    if not Path(sample_path).exists():
        print("Creating synthetic test image:", sample_path)
        create_synthetic_sample(sample_path)

    # Detect dots
    img, centers, cleaned = detect_dots_safe(sample_path, debug=True)
    print("Detected dots:", len(centers))

    if len(centers) == 0:
        print("⚠️ No dots detected. Try adjusting thresholds or use a clearer image.")
        return

    # Draw loops
    out_img = draw_loops_on_image(img, centers, out_path="kolam_loops.png")
    if out_img:
        print("Output saved to", out_img)

    # Show results
    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1); plt.imshow(cv2.cvtColor(cv2.imread(sample_path), cv2.COLOR_BGR2RGB)); plt.title("Input"); plt.axis("off")
    plt.subplot(1,3,2); plt.imshow(cleaned, cmap='gray'); plt.title("Cleaned mask"); plt.axis("off")
    if out_img:
        plt.subplot(1,3,3); plt.imshow(cv2.cvtColor(cv2.imread(out_img), cv2.COLOR_BGR2RGB)); plt.title("Kolam loops"); plt.axis("off")
    plt.tight_layout(); plt.show()

if __name__ == "__main__":
    main()
```

```python
import gradio as gr
import matplotlib.pyplot as plt
import networkx as nx
from PIL import Image
import io

def generate_kolam_graph(vertices, edges, stroke_width):
    """
    Generate a custom graph based on user-provided vertices and edges.
    """
    try:
        # Create the graph
        G = nx.Graph()
        v = int(vertices)
        G.add_nodes_from(range(1, v+1))

        # Parse edges (format: 1-2,2-3,3-1)
        edge_list = []
        if edges.strip():
            for e in edges.split(","):
                u, w = e.split("-")
                edge_list.append((int(u.strip()), int(w.strip())))
        G.add_edges_from(edge_list)

        # Draw the graph
        plt.figure(figsize=(5,5))
        pos = nx.spring_layout(G, seed=42)  # automatic layout
        nx.draw(G, pos, with_labels=True, node_color="skyblue",
                edge_color="black", width=stroke_width, node_size=700, font_size=12)

        # Save to PIL Image
        buf = io.BytesIO()
        plt.savefig(buf, format="png", bbox_inches='tight')
        buf.seek(0)
        plt.close()
        img = Image.open(buf)
        return img

    except Exception as e:
        return f"❌ Error: {str(e)}"

# ----------------- GRADIO UI -----------------
with gr.Blocks() as demo:
    gr.Markdown("## 🔗 Custom Kolam / Graph Generator using NetworkX")

    with gr.Row():
        vertices = gr.Number(value=5, label="Number of Vertices", precision=0)
        edges = gr.Textbox(value="1-2,2-3,3-1", label="Edges (format: 1-2,2-3,...)")
        stroke = gr.Slider(1, 10, value=2, step=1, label="Stroke Width")

    btn = gr.Button("Generate Graph")
    output = gr.Image(type="pil", label="Generated Graph")

    btn.click(fn=generate_kolam_graph, inputs=[vertices, edges, stroke], outputs=output)

demo.launch()
```

```python
import torch
import torch.nn as nn

# Generator
class Generator(nn.Module):
    def __init__(self, latent_dim=100):
        super(Generator, self).__init__()
        self.model = nn.Sequential(
            nn.ConvTranspose2d(latent_dim, 128, 4, 1, 0, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),

            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(True),

            nn.ConvTranspose2d(64, 1, 4, 2, 1, bias=False),
            nn.Tanh()  # Output in [-1,1]
        )
    def forward(self, z):
        return self.model(z)

# Discriminator
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(1, 64, 4, 2, 1, bias=False),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(64, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(128, 1, 4, 1, 0, bias=False),
            nn.Sigmoid()
        )
    def forward(self, img):
        return self.model(img)
```

```python
# kolam_detector.py
# Requires: pip install opencv-python numpy matplotlib

import cv2
import numpy as np
import os
from pathlib import Path
import matplotlib.pyplot as plt

# ---------------------------
# Step 1: Create Synthetic Kolam Sample (for testing)
# ---------------------------
def create_synthetic_sample(out_path="kolam_sample.jpg",
                            rows=6, cols=6, spacing=80, offset=(60,60),
                            dot_radius=6, img_size=(600,600)):
    """Create a synthetic dot-grid image for testing."""
    w, h = img_size
    bg = 255 * np.ones((h,w), dtype=np.uint8)
    for i in range(rows):
        for j in range(cols):
            x = offset[0] + j * spacing
            y = offset[1] + i * spacing
            cv2.circle(bg, (x, y), dot_radius, (0,), -1)
    cv2.imwrite(out_path, bg)
    return out_path

# ---------------------------
# Step 2: Detect Dots
# ---------------------------
def detect_dots_safe(img_path, min_area=5, max_area=5000, debug=False):
    """Detect dots (black on white) using adaptive threshold + contours."""
    if not Path(img_path).exists():
        raise FileNotFoundError(f"Image not found: {img_path}\n(Current dir: {os.getcwd()})")

    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise ValueError(f"cv2.imread failed to load image: {img_path}")

    # Invert if needed: we assume dots are dark on light background
    inv = cv2.bitwise_not(img)

    # Adaptive threshold (robust to lighting)
    thr = cv2.adaptiveThreshold(inv, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                cv2.THRESH_BINARY, 11, 3)

    # Morphological cleanup
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (3,3))
    clean = cv2.morphologyEx(thr, cv2.MORPH_OPEN, kernel)

    contours, _ = cv2.findContours(clean, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    centers = []
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area < min_area or area > max_area:
            continue
        M = cv2.moments(cnt)
        if M['m00'] == 0:
            continue
        cx = M['m10'] / M['m00']
        cy = M['m01'] / M['m00']
        centers.append((cx, cy))

    centers = np.array(centers)
    if debug:
        print(f"Contours found: {len(contours)}; Dots after filter: {len(centers)}")

    return img, centers, clean

# ---------------------------
# Step 3: Arrange Points into Grid
# ---------------------------
def arrange_grid(points, max_rows=40):
    """Arrange detected points into a grid of rows × cols."""
    if len(points) == 0:
        return None, 0, 0
    pts = np.array(points)
    N = len(pts)
    best_score = 1e9
    best_groups = None
    for r in range(1, min(N, max_rows) + 1):
        sorted_idx = np.argsort(pts[:,1])  # sort by y
        groups = np.array_split(pts[sorted_idx], r)
        if any(len(g) < 2 for g in groups):
            continue
        stds = [np.std(g[:,1]) for g in groups]
        score = np.mean(stds)
        if score < best_score:
            best_score, best_groups = score, groups
    if best_groups is None:
        best_groups = [pts[np.argsort(pts[:,1])]]
    grid = [g[np.argsort(g[:,0])] for g in best_groups]
    rows = len(grid)
    cols = int(round(np.mean([len(r) for r in grid])))
    return grid, rows, cols

# ---------------------------
# Step 4: Draw Loops Around 2×2 Cells
# ---------------------------
def draw_loops_on_image(img_gray, centers, out_path="kolam_loops.png"):
    canvas = cv2.cvtColor(img_gray, cv2.COLOR_GRAY2BGR)
    if len(centers) == 0:
        print("⚠️ No dots detected. Check threshold parameters or image quality.")
        cv2.imwrite(out_path, canvas)
        return out_path

    grid, rows, cols = arrange_grid(centers)
    max_cols = max(len(r) for r in grid)
    pts_grid = [[None]*max_cols for _ in range(len(grid))]
    for i, r in enumerate(grid):
        for j, p in enumerate(r):
            pts_grid[i][j] = (int(round(p[0])), int(round(p[1])))

    for i in range(len(grid)-1):
        for j in range(max_cols-1):
            p_tl = pts_grid[i][j]
            p_tr = pts_grid[i][j+1]
            p_bl = pts_grid[i+1][j] if i+1 < len(grid) else None
            p_br = pts_grid[i+1][j+1] if i+1 < len(grid) else None
            if None in (p_tl, p_tr, p_bl, p_br):
                continue
            minx = min(p_tl[0], p_bl[0])
            maxx = max(p_tr[0], p_br[0])
            miny = min(p_tl[1], p_tr[1])
            maxy = max(p_bl[1], p_br[1])
            center = ((minx + maxx)//2, (miny + maxy)//2)
            axes = (max(6, (maxx-minx)//2), max(6, (maxy-miny)//2))
            cv2.ellipse(canvas, center, axes, 0, 0, 360, (0,0,255), 2)

    cv2.imwrite(out_path, canvas)
    return out_path

# ---------------------------
# Step 5: Main
# ---------------------------
def main():
    sample_path = "/content/D2.jpg"
    if not Path(sample_path).exists():
        print("No sample image found -> creating a synthetic one.")
        create_synthetic_sample(sample_path)

    img, centers, cleaned = detect_dots_safe(sample_path, debug=True)
    print("Detected dots:", len(centers))

    out_img = draw_loops_on_image(img, centers, out_path="kolam_loops.png")
    print("Output saved to", out_img)

    # Show results
    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1); plt.imshow(cv2.cvtColor(cv2.imread(sample_path), cv2.COLOR_BGR2RGB)); plt.title("Input"); plt.axis("off")
    plt.subplot(1,3,2); plt.imshow(cleaned, cmap='gray'); plt.title("Thresholded"); plt.axis("off")
    plt.subplot(1,3,3); plt.imshow(cv2.cvtColor(cv2.imread(out_img), cv2.COLOR_BGR2RGB)); plt.title("Kolam Loops"); plt.axis("off")
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
